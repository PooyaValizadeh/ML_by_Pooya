```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("china_gdp.csv")
df= df.dropna()


plt.scatter(df["Year"],df["Value"])
df.dtypes
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
x= np.asanyarray(df["Year"]).reshape(-1,1)
y= np.asanyarray(df["Value"]).reshape(-1,1)
xt , xe , yt ,ye = train_test_split(x,y,test_size= 0.2)
print(xt,x)
poly = PolynomialFeatures(2)
x1 = poly.fit_transform(xt)

```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(x1,yt)
```

```python
from sklearn.metrics import r2_score
x2 = poly.fit_transform(xe)
yhat = model.predict(x2)
print(r2_score(yhat,ye))
```
