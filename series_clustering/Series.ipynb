```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Web_Series.csv")
df.head()
```

```python
df.dtypes
```

```python

df = df.join(pd.get_dummies(df["Streaming Platform"],dtype =int ))
df.head(1)
```

```python
df.columns
```

```python
x = df[df.columns.drop(['Series Title', 'Content Rating', 'IMDB Rating', 'Genre', 'Description', 'No of Seasons','Streaming Platform'])]
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.cluster import KMeans
modelkm = KMeans(n_clusters=3)
modelkm.fit(x)
```
